Trade-offs:

- Pros:
  - Lower memory churn during recursion.
  - Fewer allocations than copying on each call.
  - Standard backtracking pattern; scales better.

- Cons:
  - Easier to introduce bugs (forgetting `pop`, wrong base case, aliasing).
  - More stateful/mutation-heavy, so harder to reason about than purely functional style.

Complexity intuition:

- Backtracking buffer: `O(n)` working memory (plus output).
- Any correct solution still needs `O(n * 2^n)` total output space/time to materialize all subsets.
- Copy at leaves is unavoidable because each returned subset must be an independent object.



# object reference
Yes, in your backtracking version there is one shared `subset` list object across the active call stack unless you explicitly create a new list.

Python model (what is happening):

- Names are references: Python variables do not store object contents; they store references to objects.
- Function calls pass object references: arguments are passed by object-sharing (often described as "pass-by-assignment").
- Each call frame has its own local name bindings, but those names can point to the same object.

So for recursion:

- Frame `backtrack(i=0, subset)` has local name `subset -> S`.
- Frame `backtrack(i=1, subset)` also has local name `subset -> S`.
- Same for deeper frames, unless you create a new list.
- `append` mutates `S` in place, visible to all frames that reference `S`.
- `pop` undoes that mutation for the shared working buffer.

When is it replicated (new object created)?

- Explicit copy:
  - `subset.copy()`
  - `list(subset)`
  - `subset[:]`
- Constructing a new list expression:
  - `subset + [nums[i]]` (new list)
  - `[ *subset, nums[i] ]` (new list)
- `copy.deepcopy(subset)` creates a recursively copied structure (needed only for nested mutable elements).

Scopes vs objects:

- Scope/frame controls name visibility/lifetime (`subset` variable in each call).
- Scope does not automatically clone objects.
- Object identity is shared unless your code allocates a new object.

Quick check pattern:

```python
print(id(subset))
```

If IDs stay the same across recursive frames, all frames reference the same object.



## Python global vs local variable overriding

Python resolves names with **LEGB** order:

- **L**ocal (current function)
- **E**nclosing (outer function scopes)
- **G**lobal (module level)
- **B**uiltins (`len`, `print`, etc.)

### 1) Local assignment shadows global by default

```python
x = 10

def f():
    x = 20  # local x, does not change global x
    return x

f()   # 20
x     # 10
```

If a name is assigned anywhere in a function, Python treats it as local in that function unless declared otherwise.

### 2) Use `global` to rebind module-level variable

```python
x = 10

def f():
    global x
    x = 20  # changes module-level x

f()
x  # 20
```

### 3) Use `nonlocal` to rebind enclosing function variable

```python
def outer():
    x = 10

    def inner():
        nonlocal x
        x = 20  # changes outer x

    inner()
    return x

outer()  # 20
```

### 4) Mutating object vs rebinding name

```python
g = []

def a():
    g.append(1)   # mutate shared global object (no global keyword needed)

def b():
    global g
    g = [2]       # rebind global name to a new object (global needed)
```

- **Mutate object** (e.g., `append`, `dict[key]=...`): no `global` needed if name is only read.
- **Rebind name** (e.g., `x = ...`): `global`/`nonlocal` needed to target outer scope.

### 5) Common pitfall: `UnboundLocalError`

```python
x = 5

def f():
    print(x)  # error: x is considered local because of assignment below
    x = 6
```

Python marks `x` local for `f` due to `x = 6`, so reading it first raises `UnboundLocalError`.

```python
from typing import List


class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        ans: List[List[int]] = []
        path: List[int] = []

        def dfs(i: int) -> None:
            if i == len(nums):
                ans.append(path.copy())
                return

            # Skip nums[i]
            dfs(i + 1)

            # Take nums[i]
            path.append(nums[i])
            dfs(i + 1)
            path.pop()

        dfs(0)
        return ans
```
