````python
import random
# O(1) average design: dense array for random index access + value->index map for fast membership/removal.
class RandomizedSet:
    def __init__(self):
        self.map_value_index = {}
        self.array = []

    def insert(self, val):
        if val in self.map_value_index:
            return False
        self.map_value_index[val] = len(self.array)
        self.array.append(val)
        return True

    def remove(self, val):
        if val not in self.map_value_index:
            return False

        remove_idx = self.map_value_index[val]
        last_val = self.array[-1]

        # Move last element into removed slot (if different slot), then fix its index.
        self.array[remove_idx] = last_val
        self.map_value_index[last_val] = remove_idx

        self.array.pop()
        del self.map_value_index[val]
        return True

    def getRandom(self):
        return random.choice(self.array)
````
