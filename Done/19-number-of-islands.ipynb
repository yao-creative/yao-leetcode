### Union-Find notes: aliasing, fix, and `get_unique_count` critique

When I said **aliased**, I mean multiple rows in `self.parent` point to the **same underlying list object**.

- Current init:
  - `self.parent = [[(0,0)] * n] * m`
- Underlying behavior:
  - Python builds one inner row list, then repeats references to that same row `m` times.
  - So `self.parent[0] is self.parent[1]` is `True`.
  - Writing `self.parent[i][j] = ...` mutates that shared row for every row.

That is why your printed `final parent` often shows repeated row patterns.

Correct parent matrix construction (independent rows):

```python
self.parent = [[(i, j) for j in range(n)] for i in range(m)]
```

or equivalently:

```python
self.parent = [[(0, 0)] * n for _ in range(m)]
for i in range(m):
    for j in range(n):
        self.parent[i][j] = (i, j)
```

#### Is `get_unique_count` correct now?

Not yet. It currently tracks visited **coordinates** `(i, j)`, not visited **roots**.

- You do:
  - `if (i, j) not in unique_map: ... counter += 1 for land`
- This counts each land cell once, i.e. total land cells, not number of connected components.

Correct logic for island count:

1. For each land cell, compute `root = find((i, j))`.
2. Add `root` to a set.
3. Return `len(roots_set)`.


Two different aliasing levels are happening:

- `[(0, 0)] * n` (columns): repeats references to the same tuple object in one row.
- `[[...]] * m` (rows): repeats references to the same row list object across rows.

Why columns seem fine:

- Your column element is a tuple `(i, j)`, which is immutable.
- When you do `self.parent[i][j] = new_tuple`, Python replaces that slot with a new reference; it does not mutate the original tuple.
- So shared tuple refs are usually harmless here.

Why rows break:

- The row itself is a mutable list.
- With `* m`, all rows are the exact same list object.
- Updating one cell in one row updates that same column in every row.

So the real bug is row aliasing, not tuple aliasing.

Rule of thumb:

- Aliasing is usually dangerous with mutable objects (`list`, `dict`, `set`) because in-place mutation is shared.
- Aliasing is usually less dangerous with immutable objects (`int`, `str`, `tuple`) because values cannot be changed in place.

Underlying Python architecture (why this happens):

- Python variables and container slots store references (pointers) to `PyObject`s, not raw values.
- `*` on a list duplicates references, not deep-copies nested objects.
- For mutable targets, multiple references see the same in-place mutation.
- For immutable targets, "changes" create a new object and rebind one reference, so siblings are unaffected.

### Union by Rank (Variation) Explained

Union by rank is a strategy to keep Union-Find trees shallow during `union`.

- `rank[root]` is an upper bound on tree height (or a balancing score).
- When unioning two roots:
  - attach lower-rank root under higher-rank root;
  - if ranks are equal, choose one as parent and increment its rank by 1.

Why it matters:

- Without rank/size balancing, repeated unions can create long chains.
- Long chains make `find` slower before compression catches up.
- With union-by-rank + path compression, amortized complexity is near constant (`alpha(n)`).

Minimal pattern:

```python
def union(a, b):
    ra, rb = find(a), find(b)
    if ra == rb:
        return
    if rank[ra] < rank[rb]:
        parent[ra] = rb
    elif rank[ra] > rank[rb]:
        parent[rb] = ra
    else:
        parent[rb] = ra
        rank[ra] += 1
```

Practical note for your grid problem:

- Correctness does not require rank.
- Rank improves performance robustness on larger/adversarial merge orders.


### Where the parent write actually happens in `find`

- `x = parent[x]` is a read + local variable reassignment only.
- It does not modify the `parent` structure.
- `parent[x] = parent[parent[x]]` is a write to the parent array (path compression / path halving).

So:

- `x = parent[x]` moves the cursor upward.
- `parent[x] = ...` rewires the tree.

In your current `find`, compression does not happen unless you include a `parent[...] = ...` assignment (or recursive equivalent like `parent[x] = find(parent[x])`).


### Why the sequence diagram looked "after root", but code writes earlier

Good catch. Not a call-stack requirement; it is about which compression variant you implement.

- Diagram showed a conceptual full-compression view: once root is known, nodes on the path point to root.
- Iterative path-halving code writes during traversal:
  - `parent[x] = parent[parent[x]]` (write first)
  - `x = parent[x]` (then move)

So in path halving, rewiring starts before the final root is reached.

Recursive full compression often looks like:

```python
def find(x):
    if parent[x] != x:
        parent[x] = find(parent[x])
    return parent[x]
```

This version uses call stack and writes on unwind, which matches the "after root is known" mental model.


### Cost Trade-off: Full Compression vs Path Halving

Both give near-constant amortized performance with union-by-rank/size (`alpha(n)`), but constants differ.

Full compression (`parent[x] = find(parent[x])`):

- Flattens the path aggressively in one `find`.
- Usually fewer future hops after a call.
- Recursive style is very clear, but can add call-stack overhead (or recursion-limit risk in Python if used naively).

Path halving (`parent[x] = parent[parent[x]]` in loop):

- Compresses gradually while traversing.
- Iterative and stack-safe in Python.
- Often excellent practical throughput due to simple loop behavior, though a single call may leave slightly taller trees than full compression.

Practical guidance for Python:

- Prefer iterative halving/splitting for robustness and no recursion concerns.
- Prefer full compression when code clarity is primary and recursion depth is controlled.
- In most interview and production DSU workloads, both are effectively optimal when paired with rank/size union.

4. Escalate high-risk components with biosafety/human-in-the-loop checks.

Boundary reminder:

- DSU helps partition and deduplicate structural relationships.
- It does not replace probabilistic biological modeling, uncertainty calibration, or experimental validation.


### Generic Framework: How to Decide Whether to Use an Algorithm

Use this quick checklist before committing to any algorithm/design.

1. Problem shape
- What exact query do you need to answer (count, path, ranking, prediction, clustering)?
- Is the data static or dynamic (updates/deletes/streaming)?
- Are constraints local, global, weighted, probabilistic, or deterministic?

2. Correctness fit
- Does the algorithm's invariant exactly match the problem contract?
- Are edge cases naturally handled (empty input, duplicates, disconnected parts, adversarial inputs)?
- If not exact, is the mismatch acceptable and explicitly documented?

3. Complexity budget
- Time target: worst-case and amortized limits under expected scale.
- Space target: peak memory, intermediate buffers, and representation overhead.
- Latency mode: batch throughput vs p95/p99 online latency.

4. Data/interface constraints
- Input type/format stability (string vs int, sparse vs dense, schema drift).
- Mutability rules and concurrency model.
- Need for incremental updates vs full recomputation.

5. Operational constraints
- Debuggability: can you introspect state/invariants in prod?
- Failure modes: graceful degradation, rollback, and observability.
- Implementation risk: how likely are subtle bugs vs alternatives?

6. Decision rule (practical)
- Use the algorithm when invariant fit is exact + complexity fits budget + operational risk is acceptable.
- Avoid it when even one of these fails and no mitigation exists.
- If uncertain, prototype two candidates and compare on representative workloads.

7. Constraint table template

| Dimension | Required | Candidate A | Candidate B | Notes |
|---|---|---|---|---|
| Correctness invariant fit |  |  |  |  |

8. Red flags (usually means "do not use")
- Algorithm solves a different query than your product actually needs.
- Works only on toy scale; violates latency/memory at real scale.
- Requires assumptions your data pipeline cannot guarantee.
- Hard to validate/monitor, with high silent-failure risk.


### Union-Find Family: When to Use Which Variant

Core fit test (for any DSU/Union-Find variant):

- Your query is connectivity/component membership (same set? component count?).
- Graph updates are mostly edge additions (monotonic connectivity).
- You need very fast repeated unions/finds.

If these are true, DSU is usually a strong choice.

Union-Find family and trade-offs:

- Quick-Find:
  - `find` is O(1), `union` is O(n).
  - Use only for tiny n or teaching.
- Quick-Union:
  - naive trees, can degrade badly.
  - Usually not used alone in practice.
- Weighted/Rank Union:
  - keeps trees shallow by attaching smaller/lower-rank root.
  - baseline practical improvement.
- Path Compression (full):
  - aggressive flattening during `find`.
  - excellent amortized performance.
- Path Halving/Splitting (iterative compression variants):
  - compress during traversal without recursion.
  - often preferred in Python for stack safety and simple loops.
- Rollback DSU:
  - supports undo of unions (offline dynamic connectivity patterns).
  - useful when you need reversible state across time segments.
- Persistent/Versioned DSU:
  - query historical versions.
  - higher implementation complexity, used in specialized workflows.

When to use DSU (practical constraints):

- IDs are stable/indexable.
- No frequent deletions needed.
- Need high-throughput component queries over large workloads.

When NOT to use DSU:

- Frequent edge deletions (use dynamic graph structures instead).
- Need shortest paths, top-k routes, centrality, flow, or ranking.
- Relations are weighted/probabilistic where thresholding changes often.

Variant picker (default recommendation):

- Default: union-by-rank/size + path halving (iterative).
- If recursion safe and you want clarity: union-by-rank/size + full compression.
- If time-travel/undo is required: rollback or persistent DSU variant.

Quick decision checklist:

1. Is the task connectivity, not path optimization?
2. Are updates mostly unions, not deletions?
3. Do I need repeated component queries at scale?
4. Can I map elements to compact IDs safely?

If 1-4 are yes, choose a DSU variant; otherwise choose a graph/search/ranking method.
