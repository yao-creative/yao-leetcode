```python
class FreqStack:
    def __init__(self):
        pass

    def push(self, val: int) -> None:
        pass

    def pop(self) -> int:
        pass
```

```python
def test(solution):
    cases = [
        (((["push","push","push","push","push","push","pop","pop","pop","pop"], [[5],[7],[5],[7],[4],[5],[],[],[],[]])), [None,None,None,None,None,None,5,7,5,4]),
    ]
    for i, (args, expected) in enumerate(cases, 1):
        got = solution(*args)
        assert got == expected, f'case {i}: expected {expected}, got {got}'

```

```python
def current_solution(ops, params):
    obj = FreqStack()
    out = []
    for op, arg in zip(ops, params):
        out.append(getattr(obj, op)(*arg))
    return out

test(current_solution)
print("PASS")
```
