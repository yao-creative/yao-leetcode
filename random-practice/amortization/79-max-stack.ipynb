```python
class MaxStack:
    def __init__(self):
        pass

    def push(self, x: int) -> None:
        pass

    def pop(self) -> int:
        pass

    def top(self) -> int:
        pass

    def peekMax(self) -> int:
        pass

    def popMax(self) -> int:
        pass
```

```python
def test(solution):
    cases = [
        (((["push","push","push","top","popMax","top","peekMax","pop","top"], [[5],[1],[5],[],[],[],[],[],[]])), [None,None,None,5,5,1,5,1,5]),
    ]
    for i, (args, expected) in enumerate(cases, 1):
        got = solution(*args)
        assert got == expected, f'case {i}: expected {expected}, got {got}'

```

```python
def current_solution(ops, params):
    obj = MaxStack()
    out = []
    for op, arg in zip(ops, params):
        out.append(getattr(obj, op)(*arg))
    return out

test(current_solution)
print("PASS")
```
