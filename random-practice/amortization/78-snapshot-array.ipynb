```python
class SnapshotArray:
    def __init__(self, length: int):
        pass

    def set(self, index: int, val: int) -> None:
        pass

    def snap(self) -> int:
        pass

    def get(self, index: int, snap_id: int) -> int:
        pass
```

```python
def test(solution):
    cases = [
        (((3, ["set","snap","set","get"], [[0,5],[],[0,6],[0,0]])), [None,0,None,5]),
    ]
    for i, (args, expected) in enumerate(cases, 1):
        got = solution(*args)
        assert got == expected, f'case {i}: expected {expected}, got {got}'

```

```python
def current_solution(length, ops, params):
    obj = SnapshotArray(length)
    out = []
    for op, arg in zip(ops, params):
        out.append(getattr(obj, op)(*arg))
    return out

test(current_solution)
print("PASS")
```
