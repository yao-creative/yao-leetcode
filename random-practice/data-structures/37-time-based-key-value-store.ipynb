```python
Problem: Time Based Key-Value Store
Difficulty: Medium
Link: https://leetcode.com/problems/time-based-key-value-store/

Example:
set("foo", "bar", 1)
get("foo", 1) -> "bar"
get("foo", 3) -> "bar"

Constraints:
- Up to 2 * 10^5 calls to set/get
```

```python
class TimeMap:
    def __init__(self):
        pass

    def set(self, key: str, value: str, timestamp: int) -> None:
        pass

    def get(self, key: str, timestamp: int) -> str:
        pass

```

```python
def current_solution(ops, params):
    tm = TimeMap()
    out = []
    for op, arg in zip(ops, params):
        if op == "set":
            out.append(tm.set(*arg))
        else:
            out.append(tm.get(*arg))
    return out

result = "PASS (No solution provided to execute)"
print(result)
# When TimeMap is runnable, replace the two lines above with:
# test(current_solution)
# print("PASS")

```
