### Problem: Design Ad Eligibility Pool
**Difficulty:** Medium  
**Link:** Practice variant inspired by LeetCode 380  
**Tags:** Hash Map, Array, Ad Serving

**Example:**
`["AdPool","addAd","addAd","drawAd","removeAd","drawAd"] / [[],[101],[202],[],[101],[]] -> [null,true,true,101|202,true,202]`

**Constraints:**
- Average O(1) for `addAd`, `removeAd`, `drawAd`
- `drawAd` is uniform over currently eligible ads
- No duplicate eligible ads


```python
import random

class AdPool:
    def __init__(self):
        pass

    def addAd(self, ad_id: int) -> bool:
        pass

    def removeAd(self, ad_id: int) -> bool:
        pass

    def drawAd(self) -> int:
        pass
```

```python
def test(solution):
    ops = ["AdPool", "addAd", "addAd", "drawAd", "removeAd", "addAd", "drawAd"]
    params = [[], [101], [202], [], [101], [202], []]

    got = solution(ops, params)

    assert got[0] is None
    assert got[1] is True
    assert got[2] is True
    assert got[3] in {101, 202}
    assert got[4] is True
    assert got[5] is False
    assert got[6] == 202
```

```python
def current_solution(ops, params):
    obj = None
    out = []
    for op, arg in zip(ops, params):
        if op == "AdPool":
            obj = AdPool(*arg)
            out.append(None)
            continue
        out.append(getattr(obj, op)(*arg))
    return out

# Uncomment after implementation:
# test(current_solution)
# print("PASS")
```
