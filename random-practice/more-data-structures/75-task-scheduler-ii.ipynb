
Problem: Task Scheduler II

Difficulty: Medium

Link: https://leetcode.com/problems/task-scheduler-ii/

Example:
Input: tasks = [1,2,1,2,3,1], space = 3 -> Output: 9

Constraints:
- 1 <= tasks.length <= 10^5
- 1 <= tasks[i] <= 10^9
- 1 <= space <= tasks.length


```python
from typing import List
import heapq 

class Solution:
    def taskSchedulerII(self, tasks: List[int], space: int) -> int:
    
        # we can use an space-sized-heap that we can first read the whole input, 
        # we don't have to read and write at the same time, since we need minimal read all of the input complexity
        # I was going to store count x interval, however interval is standard for all tasks so we can just store number within the heap
        
        init_dict = {}
        for task in tasks:
            if task not in init_dict:
                init_dict[task] = 1
            else:
                init_dict[task] += 1
        heap = [(-v, k) for k, v in init_dict.items()]
        heapq.heapify(heap)

        last_dict = {k: -1 for k in init_dict.keys()} # key: task, value: last_position.
        #normally we place in order greedy, maximum time left which is basically interval x count first. 
        # ** If two values are tied at count using direct index replacement A for B gives a same length task scheduling solution.
        # This is minimal since, invariant is that we are minimizing the maximum time left. 
        #how to figure out spacing? we can count by last position the task was placed cacheing  
        snapshot = []
        cooldown = []
        heapq.heapify(cooldown)
        # technically can also be heap, storing the next valid index and if we reach there we add it back to the original heap
        i = 0
        while heap or cooldown:
            if cooldown: #try to pop and put back to heap until can't anymore
                # print(f"cooldown: {cooldown}")
                while cooldown and cooldown[0][0] <= i: #which means valid from cooldown indices
                    count, task = heapq.heappop(cooldown)[1]
                    heapq.heappush(heap,(count, task))
            if not heap and cooldown and cooldown[0][0] > i: #none left to attempt in the heap cooling down earliest index not yet.
                snapshot.append("idle")
                i += 1
                # else, heap is not empty (heap) then we check the next in the heap without incrementing or we're done technically (not heap and not cooldown)
            else: #heap
                count, task = heapq.heappop(heap) #take the furthest count
                # print(f"i: {i}, count: {count}, task: {task}")
                last_seen =  last_dict[task]
                if last_seen == -1 or i - last_seen > space: #accept
                    last_dict[task] = i 
                    count += 1
                    if count < 0:
                        heapq.heappush(heap, (count, task))
                    snapshot.append(task)
                    i += 1

                else:
                    # Not accepted, thus add tentative to cooldown

                    # add tentative to cooldown, 
                    cooldown_tup = (last_seen + space + 1 ,(count, task))
                    # print(f"cooldown_tup: {cooldown_tup}")
                    heapq.heappush(cooldown, cooldown_tup)
                    

                
                #pushback, technically heap will preserve ordering and due to **, 
                # we are guaranteed shortest solution by each pop from heap
            # print(f"snapshot: {snapshot}")

            #technically this scheduling pushes all of the runs towards the end, 
            # but if we stored it in a deque and iterated the other direction popping,
            # we can get frontloaded scheduling
        return i

            
        
```
