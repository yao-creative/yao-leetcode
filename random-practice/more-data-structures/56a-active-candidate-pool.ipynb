### Problem: Design Active Candidate Pool
**Difficulty:** Medium  
**Link:** Practice variant inspired by LeetCode 380  
**Tags:** Hash Map, Array, Randomized Design

**Example:**
`["ActivePool","activate","activate","pick","deactivate","pick"] / [[],[10],[20],[],[10],[]] -> [null,true,true,10|20,true,20]`

**Constraints:**
- `activate`, `deactivate`, and `pick` should be average O(1)
- `pick` is called only when pool is non-empty
- Each active element must be returned with equal probability


```python
import random

class ActivePool:
    def __init__(self):
        pass

    def activate(self, candidate_id: int) -> bool:
        """Return True if newly activated, False if already active."""
        pass

    def deactivate(self, candidate_id: int) -> bool:
        """Return True if removed, False if not active."""
        pass

    def pick(self) -> int:
        """Return a uniformly random active candidate."""
        pass
```

```python
def test(solution):
    ops = ["ActivePool", "activate", "activate", "pick", "deactivate", "activate", "pick"]
    params = [[], [10], [20], [], [10], [20], []]

    got = solution(ops, params)

    assert got[0] is None
    assert got[1] is True
    assert got[2] is True
    assert got[3] in {10, 20}
    assert got[4] is True
    assert got[5] is False
    assert got[6] == 20
```

```python
def current_solution(ops, params):
    obj = None
    out = []
    for op, arg in zip(ops, params):
        if op == "ActivePool":
            obj = ActivePool(*arg)
            out.append(None)
            continue
        out.append(getattr(obj, op)(*arg))
    return out

# Uncomment after implementation:
# test(current_solution)
# print("PASS")
```
