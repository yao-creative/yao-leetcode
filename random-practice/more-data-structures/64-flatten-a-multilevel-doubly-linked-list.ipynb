```python
class Node:
    def __init__(self, val, prev=None, next=None, child=None):
        self.val = val
        self.prev = prev
        self.next = next
        self.child = child

class Solution:
    def flatten(self, head):
        pass

```

```python
def build_sample(values, child_map):
    nodes = [Node(v) for v in values]
    for i in range(len(nodes) - 1):
        nodes[i].next = nodes[i + 1]
        nodes[i + 1].prev = nodes[i]
    for parent_idx, child_values in child_map.items():
        child_head = build_sample(child_values, {})
        nodes[parent_idx].child = child_head
    return nodes[0] if nodes else None

def to_list(head):
    out = []
    while head:
        out.append(head.val)
        assert head.child is None, f'child pointer not cleared at {head.val}'
        head = head.next
    return out

def test(solution):
    cases = [
        (((( [1, 2, 3], {1: [7, 8]}),), [1, 2, 7, 8, 3])),
        (((( [1, 2], {}),), [1, 2])),
    ]
    for i, (args, expected) in enumerate(cases, 1):
        got = solution(*args)
        assert got == expected, f'case {i}: expected {expected}, got {got}'

```
