### Problem: Design Agent Tool Router Set
**Difficulty:** Medium  
**Link:** Practice variant inspired by LeetCode 380  
**Tags:** Hash Map, Array, Randomized Exploration

**Example:**
`["ToolRouter","enable","enable","sampleTool","disable","sampleTool"] / [[],[1],[2],[],[1],[]] -> [null,true,true,1|2,true,2]`

**Constraints:**
- `enable`, `disable`, `sampleTool` average O(1)
- No duplicate active tools
- `sampleTool` must be uniform over currently enabled tools


```python
import random

class ToolRouter:
    def __init__(self):
        pass

    def enable(self, tool_id: int) -> bool:
        pass

    def disable(self, tool_id: int) -> bool:
        pass

    def sampleTool(self) -> int:
        pass
```

```python
def test(solution):
    ops = ["ToolRouter", "enable", "enable", "sampleTool", "disable", "enable", "sampleTool"]
    params = [[], [1], [2], [], [1], [2], []]

    got = solution(ops, params)

    assert got[0] is None
    assert got[1] is True
    assert got[2] is True
    assert got[3] in {1, 2}
    assert got[4] is True
    assert got[5] is False
    assert got[6] == 2
```

```python
def current_solution(ops, params):
    obj = None
    out = []
    for op, arg in zip(ops, params):
        if op == "ToolRouter":
            obj = ToolRouter(*arg)
            out.append(None)
            continue
        out.append(getattr(obj, op)(*arg))
    return out

# Uncomment after implementation:
# test(current_solution)
# print("PASS")
```
