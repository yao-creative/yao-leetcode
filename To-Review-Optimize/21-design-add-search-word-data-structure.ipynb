````python
class Node:
    def __init__(self, value = "", count = 0, children = None):
        self.value = value
        self.count = count
        self.children = dict()
    
    def __repr__(self):
        return f"Node(value: {self.value}, count: {self.count}, children: {[c for c in self.children]})"


class WordDictionary:
    def __init__(self):
        self.root = Node()

    def addWord(self, word: str) -> None:
        # traverse the word if it's in the tree
        # print(f"adding: {word}")
        node = self.root
        for c in word:
            # print(f"node: {node}")
            children_vals = {child_val : child for child_val, child in node.children.items()}
            # print(f"children_vals: {children_vals}")
            if c not in children_vals:
                # print(f"adding character {c}")
                node.children[c] = Node(c)
            
            node = node.children[c]
        #add counter to last node that there's a word here
        node.count += 1
        # print(f"final node: {node}")
    
    def wildcard(self, word, i, node): #already seen a first "." so enter, i = pos of first "."
        #collect all children nodes for a bounded DFS.
        stack = [(1, subnodes) for subnodes in node.children.values()]
        
        wildcard_depth = 0
        while i + wildcard_depth < len(word) and word[i + wildcard_depth] == ".":
            wildcard_depth +=1 

        # final i + wildcard_depth will point to the last "."
        
        #end case partitions: either ends with letter or "." is last char of word
        if i + wildcard_depth == len(word): 
            # if end of word then we'd like the character of wildcard_depth to have count > 0
            while len(stack) > 0:
                curr = stack.pop()
                if curr[0] > wildcard_depth: #too far ignore
                    continue
                elif curr[0] == wildcard_depth:
                    if curr[1].count > 0: #word found
                        return True 
                    #else also continue this one didn't count
                else: # less than wildcard depth keep exploring
                    stack.extend([(curr[0] + 1, child) for child in curr[1].children.values()])
            return False #reached the end of the word but no word counts found
        else: 
            # if ends in character word[i], then we have to depth first search word[i + wildcard_depth + 1]
            # print(f"i: {i} wildcard_depth: {wildcard_depth},  word[i + wildcard_depth]: { word[i + wildcard_depth]}")
            # print(f"stack: {stack}")
            target = word[i + wildcard_depth]
            while len(stack) > 0:
                curr = stack.pop()
                if curr[0] > wildcard_depth: #too far ignore
                    continue
                elif curr[0] == wildcard_depth:
                    # print(f"checking same as depth nodes curr[1]: {curr[1]}")
                    if target in curr[1].children: #target character to continue is found.
                        return i + wildcard_depth, curr[1].children[target] 
                    #else also continue this one didn't count
                else: # less than wildcard depth keep exploring
                    stack.extend([(curr[0] + 1, child) for child in curr[1].children.values()])
            return False






        

    def search(self, word: str) -> bool:
        node = self.root
        # print(f"searching: {word}")
        i = 0
        while i < len(word):
            # print(f"node: {node}")
            children_vals = {child_val : child for child_val, child in node.children.items()}
            # dfs wildcard.
            c = word[i]
            if c == ".": #wild card iteration
                wildcard_search = self.wildcard(word, i, node)
                # print(f"wildcard_search res: {wildcard_search}")
                if wildcard_search is True:
                    return True
                elif wildcard_search is False:
                    return False
                else:
                    i, node = wildcard_search #update index and node to search
            elif c in children_vals:
                node = node.children[c]
            else: # add char to children
                return False
            i +=1
        # print(f"final node: {node}")
        if node.count > 0:
            return True
        else:
            return False



````

### How to tell which mutable defaults are shared in Python

  Use this rule:

  - If the mutable object is created **once at definition time**, it is **shared**.
  - If it is created **inside the function/method call path**, it is **new each time** (not shared).

  ---

  #### 1) Function default arguments: **shared**
  ```python
  def f(x, acc=[]):   # acc is created once when def runs
      acc.append(x)
      return acc

  acc is reused across calls.

  Safe pattern:

  def f(x, acc=None):
      if acc is None:
          acc = []
      acc.append(x)
      return acc

  class Node:
      children = []   # one list shared by all Node objects

  Per-instance version:

  class Node:
      def __init__(self):
          self.children = []

```python
class Node:
    __slots__ = ("children", "is_end")

    def __init__(self):
        self.children = {}
        self.is_end = False


class WordDictionary:
    def __init__(self):
        self.root = Node()

    def addWord(self, word: str) -> None:
        node = self.root
        for ch in word:
            if ch not in node.children:
                node.children[ch] = Node()
            node = node.children[ch]
        node.is_end = True

    def search(self, word: str) -> bool:
        def dfs(i: int, node: Node) -> bool:
            if i == len(word):
                return node.is_end

            ch = word[i]
            if ch == ".":
                for child in node.children.values():
                    if dfs(i + 1, child):
                        return True
                return False

            nxt = node.children.get(ch)
            if nxt is None:
                return False
            return dfs(i + 1, nxt)

        return dfs(0, self.root)
```

Why this is algorithmically optimal for this interface:
- You cannot asymptotically avoid wildcard branch exploration in the worst case because the query semantics require checking potentially many branches.
- The practical optimization is aggressive pruning + early exit, which this version does.

Optional advanced optimization (only for heavy wildcard workloads):
- Maintain a word-length index (`length -> exists`) to fast-fail impossible lengths before DFS.

- Cache `(node_id, i)` for repeated subproblems only if your trie supports DAG-like sharing or patterns are reused in batched search; in pure trie with unique parent chains this is often low value.
