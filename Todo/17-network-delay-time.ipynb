```python
from copy import deepcopy


def network_delay_time_test_cases():
    return [
        ((([[2, 1, 1], [2, 3, 1], [3, 4, 1]], 4, 2)), 2),
        ((([[1, 2, 1]], 2, 1)), 1),
        ((([[1, 2, 1]], 2, 2)), -1),
        ((([[1, 2, 5]], 2, 1)), 5),
        ((([[1, 2, 1], [2, 3, 2], [1, 3, 4]], 3, 1)), 3),
        ((([[1, 2, 1], [2, 1, 3]], 3, 1)), -1),
    ]


def run_network_delay_time_harness(fn):
    for idx, (args, expected) in enumerate(network_delay_time_test_cases()):
        times, n, k = deepcopy(args)
        got = fn(times, n, k)
        assert got == expected, (
            f"Case {idx} failed: expected={expected}, got={got}, "
            f"times={times}, n={n}, k={k}"
        )
    return f"PASS ({len(network_delay_time_test_cases())} cases)"
```
