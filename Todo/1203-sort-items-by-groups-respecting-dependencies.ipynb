# 1203. Sort Items by Groups Respecting Dependencies

There are `n` items and `m` groups. `group[i]` is the group of item `i`, or `-1` if it belongs to no group yet. Return any ordering such that items from the same group appear contiguously and every dependency in `beforeItems` is respected. Return `[]` if impossible.

Reference examples:
- `n = 8, m = 2, group = [-1,-1,1,0,0,1,0,-1], beforeItems = [[],[6],[5],[6],[3,6],[],[],[]]` -> one valid answer is `[6,3,4,1,5,2,0,7]`
- `n = 8, m = 2, group = [-1,-1,1,0,0,1,0,-1], beforeItems = [[],[6],[5],[6],[3],[],[4],[]]` -> `[]`

Constraints worth noticing:
- `1 <= m <= n <= 3 * 10^4`.
- Ungrouped items (`-1`) usually need to be assigned their own unique synthetic groups.
- This is a two-level topological sort problem: groups, then items within groups.


```python
def is_valid_order(n, group, beforeItems, order):
    if len(order) != n or sorted(order) != list(range(n)):
        return False
    pos = {x: i for i, x in enumerate(order)}
    for i in range(n):
        for prev in beforeItems[i]:
            if pos[prev] > pos[i]:
                return False
    seen_groups = []
    for item in order:
        g = group[item]
        if g != -1 and (not seen_groups or seen_groups[-1] != g):
            seen_groups.append(g)
    return len(seen_groups) == len(set(seen_groups))


def test(solution):
    cases = [
        ((8, 2, [-1, -1, 1, 0, 0, 1, 0, -1], [[], [6], [5], [6], [3, 6], [], [], []]), 'valid'),
        ((8, 2, [-1, -1, 1, 0, 0, 1, 0, -1], [[], [6], [5], [6], [3], [], [4], []]), []),
        ((5, 5, [0, 1, 2, 3, 4], [[], [0], [1], [2], [3]]), [0, 1, 2, 3, 4]),
    ]

    for i, (args, expected) in enumerate(cases, 1):
        got = solution(*args)
        if expected == 'valid':
            assert is_valid_order(args[0], args[2], args[3], got), f"case {i}: invalid order {got}"
        else:
            assert got == expected, f"case {i}: expected={expected}, got={got}"
```
