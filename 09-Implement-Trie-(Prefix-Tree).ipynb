```python
class TrieNode:
    def __init__(self, value, wordCount):
        self.children = {}
        self.value = value
        self.wordCount = wordCount
    def __repr__(self):
        return f"TrieNode({self.children}, {self.value}, {self.wordCount})"


class Trie:

    def __init__(self):
        self.root = TrieNode(None,0)

    def insert(self, word: str) -> None:
        current_node = self.root
        for i,c in enumerate(word):
            # if in children go down children
            if c in current_node.children: 
                current_node = current_node.children[c]
            else:  #initialize and then go down
                current_node.children[c] = TrieNode(c,0)
                current_node = current_node.children[c]
            if i == len(word) -1:
                current_node.wordCount +=1 


    def search(self, word: str) -> bool:
        current_node = self.root
        for c in (word):
            print(current_node)
            # if in children go down children
            if c in current_node.children: 
                current_node = current_node.children[c]
            else: #else it is broken some where 
                return False
        if current_node.wordCount > 0:
            return True
        else: 
            return False
        
    def startsWith(self, prefix: str) -> bool:
        
        current_node = self.root
        for c in (prefix):
            # if in children go down children
            print(current_node)
            if c in current_node.children: 
                current_node = current_node.children[c]
            else: #else it is broken some where 
                return False
        return True 

# Your Trie object will be instantiated and called as such:
# obj = Trie()
# obj.insert(word)
# param_2 = obj.search(word)
# param_3 = obj.startsWith(prefix)
```
